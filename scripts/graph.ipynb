```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('serialization_results.csv', delimiter=';')

# Removing the anomaly code (if any)
df = df.dropna()

# Keep only 2 decimals for data size
df['Data Size'] = df['Data Size']

# Extracting data columns
data_size = df['Data Size']
java_serialization = df['Java Serialization']
json_serialization = df['JSON Serialization']
serenity_serialization = df['Serenity Serialization']

# Plotting the graph
plt.figure(figsize=(15, 8))

# Plotting lines for each serialization method
plt.plot(data_size, java_serialization, label='Java Serialization', marker='o')
plt.plot(data_size, json_serialization, label='JSON Serialization', marker='o')
plt.plot(data_size, serenity_serialization, label='Serenity Serialization', marker='o')

# Adding labels and title
plt.xlabel('Data Size (KB)')
plt.ylabel('Time Taken (Micro sec)')
plt.title('Comparison of Serializations')
plt.legend()

# Displaying the graph
plt.grid(True)
plt.tight_layout()
plt.show()
```
